```python
fmri["event"].unique()
```

```python
sns.relplot(x="timepoint",y="signal",data=fmri,kind="line",ci="sd")
```

```python
sns.relplot(x="timepoint",y="signal",data=fmri,kind="line",hue="region",style="event")
```

# Bar plot

```python
import matplotlib.pyplot as plt
```

```python
titanic=sns.load_dataset("titanic")
```

```python
titanic
```

```python
sns.catplot(x="sex",y="survived",kind="bar",data=titanic)
```

```python
sns.catplot(x="sex",y="survived",hue="class",kind="bar",data=titanic)
```

```python
sns.catplot(x="deck",kind="count",palette="ch:0.4",data=titanic)
```

```python
sns.catplot(y="deck",kind="count",edgecolor="green",palette="pastel",data=titanic,hue="class")
```
