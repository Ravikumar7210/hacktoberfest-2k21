```python
import pandas as pd
```

```python
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
Foods = pd.read_csv("Restaurant - Foods.csv")
```

```python
week1.head()
```

```python
week1.merge(week2,how='inner',on=["Customer ID"],suffixes=[" week 1"," week 2"])
```

```python
data = week1.merge(week2,how='outer',on=['Customer ID'],indicator=True,suffixes=[' week 1',' week 2'])
```

```python
data['_merge'].value_counts()
```

```python
week1.merge(week2,how="inner",left_index=True,right_index=True,suffixes=[" week 1"," week 2"])
```

# Group By

```python
import pandas as pd
df = pd.read_csv("fortune1000.csv")
df
```

```python
sector = df.groupby("Sector")
```

```python
sector.sum()
```
